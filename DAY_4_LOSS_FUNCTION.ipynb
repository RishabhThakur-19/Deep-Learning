```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn


df=pd.read_csv('./DATASETS/placement.csv')
print(df.shape)
print(df.head())

plt.figure(figsize=(10,6))
plt.scatter(df['cgpa'],df['iq'],c=df['placement'],cmap="winter")

x=df[['cgpa','iq']].to_numpy()
y=df[['placement']].to_numpy()

def perceptron(x,y):
    w1=w2=b=1
    lr=0.1
    for i in range(1000000):
        for j in range(x.shape[0]):
            z=w1*x[j][0]+w2*x[j][1]+b
            y_new=step(y[j])
            # print(z)
            if z*y_new<0:
                w1=w1+lr*(y_new)*x[j][0]
                w2=w2+lr*(y_new)*x[j][1]
                b=b+lr*(y_new)
    return w1,w2,b 

def step(z):
    if z<=0:
        return -1
    else:
        return 1

w1,w2,b=perceptron(x,y)
print(w1,w2,b)





m=-w1/w2
c=-b/w2
x1=np.linspace(3,10,100)
y1=m*x1+c
plt.figure(figsize=(10,6))
plt.plot(x1,y1,'red',linewidth=3)
plt.scatter(df['cgpa'],df['iq'],c=df['placement'],cmap="winter")
plt.ylim(0,300)
```
