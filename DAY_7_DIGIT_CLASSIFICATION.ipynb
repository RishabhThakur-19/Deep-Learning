```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense ,Flatten

(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()

                                      #MNIST DATASET IS ALREADY IN THE KERAS LIBRABRY AS IT IS VERY FAMOUS
                                      # KERAS HAD ALREADY  DIVIDED THE DATASET INTO TRAIN AND TEST DATA AS CAN BE SEEN ABOVE

x_train.shape                                                     #(60000,28,28)
x_train[0].shape                                                  #(28,28)
print(x_train[7])
print(plt.imshow(x_train[0]))             # used to plot 2d array data in graph

                                          #STANDARDIZING THE DATASET VALUE MAKING ALL THE VALUES IN THE RANGE OF (0-1)
                                          # WE CAN DO IT BY JUST DIVIDING IT BY 256  as max value is 256 AND MINIMUN VALUE IS 0
x_train=x_train/256
x_test=x_test/256

                                            #NOW TO WE HAVE CREATE THE MODEL
```

```python
#CREATING THE MODEL
model=Sequential()

#WE NEED TO GIVE 784 INPUTS BUT OUR DATA IS IN THE FORM OF 2D ARRAY [28][28] 
# SO WE HAVE TO CONVERT THE 2D ARRAY INTO 1D ARRAY SO WE CAN GIVE IT TO THE INPUT
# FOR THIS WE USE FLATTEN FUNCTION 

model.add(Flatten(input_shape=(28,28)))  #ADDING FLATTEN LAYER TO CONVERT 2D ARRAY INTO 1D ARRAY
model.add(Dense(128,activation="relu"))   #NO NEEDD TO SPECIFY INPUT DIM AS FLATTEN WILL HANDLE IT
model.add(Dense(64,activation="relu"))   #NO NEEDD TO SPECIFY INPUT DIM AS FLATTEN WILL HANDLE IT
model.add(Dense(10,activation="softmax"))  #OUTPUT LAYER , WHENEVER THE OUTPUT IS MORE THAN 1 CLASS WE USE SOFTMAX ACTIVATION FUNCTION
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer='adam',metrics=["accuracy"])
history=model.fit(x_train,y_train,epochs=30,validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
n_prob=model.predict(x_test)

y_prob=y_prob.argmax(axis=1)
print(y_prob)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_prob)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
```
