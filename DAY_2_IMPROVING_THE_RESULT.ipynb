```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn


df=pd.read_csv('./DATASETS/placement.csv')
print(df.shape)
print(df.head())

plt.figure(figsize=(10,6))
plt.scatter(df['cgpa'],df['iq'],c=df['placement'],cmap="winter")

x=df[['cgpa','iq']].to_numpy()
y=df[['placement']].to_numpy()
# print(x)
# print(len(x))
# print(y)

def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    # print(x)
    weights=np.ones(x.shape[1])
    lr=1
    i=0
    for i in range(1000000):
       j=np.random.randint(0,100)
       y_hat=step(np.dot(x[j],weights))
       sum=(lr*(y[j]-y_hat)*x[j])
       weights=weights+sum
    return weights[0],weights[1:] 

def step(z):
    if z>=0:
        return 1
    else:
        return 0  

intercept_,coef_=perceptron(x,y) 
print(intercept_,coef_)   


m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])

x1=np.linspace(2,10,1000)
y1=m*x1+b

plt.figure(figsize=(10,6))
plt.plot(x1,y1,color='red',linewidth=3)
plt.scatter(df['cgpa'],df['iq'],c=df['placement'],cmap="winter")
plt.ylim(-10,350)
```
